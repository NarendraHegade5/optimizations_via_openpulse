```python
import numpy as np
import time
from collections import namedtuple
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import qiskit as q
import qiskit.tools.jupyter
from qiskit.pulse import pulse_lib as _  # only want to call pulse_lib as q.pulse.pulse_lib
```

```python
samples = q.pulse.pulse_lib.discrete.drag(duration=160, amp=-0.3, sigma=35, beta=-3.5).samples
q.pulse.SamplePulse(np.concatenate([samples, samples])).draw(
    style=q.visualization.pulse.qcstyle.PulseStyle(dpi=1000),
    filename='standard_x.svg')
```

```python
q.pulse.SamplePulse(np.concatenate([samples * -2, [0] * 160])).draw(
    style=q.visualization.pulse.qcstyle.PulseStyle(dpi=1000), filename='direct_x.svg')
```
