```python
import qutip
from qutip import *
import numpy as np
```

```python
b = Bloch()
```

```python
b.add_states([basis(2, 0)])
b.add_states([basis(2, 1)])
b.add_states([(2*basis(2, 0) + np.exp(1j*1.4)*basis(2, 1)).unit()])
```

```python
b.show()
```

```python
b.save('bloch_states.svg')
# I edited it slightly in Preview and saved it as bloch_states.png
```
